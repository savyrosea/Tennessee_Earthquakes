```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

pd.set_option('display.max_rows', 1500)
```

```python
#set url
url = 'https://en.wikipedia.org/wiki/List_of_deadly_earthquakes_since_1900'
```

```python
#creating request
r = requests.get(url)
```

```python
#receiving request as text
html_doc = r.text
```

```python
soup = BeautifulSoup(html_doc)
```

```python
pretty_soup = soup.prettify()
```

```python
pretty_soup[:1000]
```

```python
soup.find('table')
```

```python
soup.find('table', attrs={'class' : 'wikitable'})
```

```python
table_html = str(soup.find('table', attrs={'class' : 'wikitable'}))
```

```python
table = pd.read_html(table_html)
```

```python
wikipedia_df = table[0]
```

```python
wikipedia_df.head()
```

```python
columns = ['origin_utc', 'country',
       'lat', 'long', 'depth_km', 'magnitude', 'secondary_effects',
       'pde_shaking_deaths', 'pde_total_deaths', 'utsu_total_deaths',
       'em_dat_total_deaths', 'other_source_deaths']
```

```python
wikipedia_df.columns = columns
```

```python
wikipedia_df.head()
```

```python
wikipedia_df.info()
```

```python
#change origin column to datetime
wikipedia_df['origin_utc'] = pd.to_datetime(wikipedia_df['origin_utc'])
```

```python
#lat and long are not something we need in the dataset, so we'll leave them as strings unless needed later
```

```python
#the ? is causing errors when trying to convert data types
wikipedia_df[wikipedia_df['depth_km'] == '?']
```

```python
wikipedia_df.iloc[413] = wikipedia_df.iloc[413].replace('?','NaN')
```

```python
wikipedia_df.iloc[413] = wikipedia_df.iloc[413].replace('??', 'NaN')
```

```python
wikipedia_df['lat'] = wikipedia_df['lat'].astype('float')
```

```python
wikipedia_df['long'] = wikipedia_df['long'].astype('float')
```

```python
wikipedia_df['depth_km'] = wikipedia_df['depth_km'].astype('float')
```

```python
#converted to a string with as type and then said we want the first 3 elements of the string
wikipedia_df['magnitude'] = wikipedia_df['magnitude'].astype(str).str[0:3]
```

```python
# astype float fails and this finds where 5 M comes from so we can fix individually
wikipedia_df[wikipedia_df['magnitude'] == '5 M']
```

```python
wikipedia_df.iloc[439]
```

```python
wikipedia_df.iloc[439] = wikipedia_df.iloc[439].replace('5 M', '5.0')
```

```python
#converting to floats should work now that 439 is fixed
wikipedia_df['magnitude'] = wikipedia_df['magnitude'].astype('float')
```

```python
wikipedia_df.info()
```

```python
wikipedia_df[wikipedia_df['em_dat_total_deaths'] == '10[7]|']
```

```python
wikipedia_df.iloc[439] = wikipedia_df.iloc[439].replace('10[7]|', '10')
```

```python
wikipedia_df['em_dat_total_deaths'] = wikipedia_df['em_dat_total_deaths'].astype('float')
```

```python
wikipedia_df['other_source_deaths'].value_counts()
```

```python
#removed brackets and digits inside
wikipedia_df['other_source_deaths'] = wikipedia_df['other_source_deaths'].str.replace(r'\[\d+\]', '')
```

```python
wikipedia_df['other_source_deaths'].value_counts()
```

```python
#removes + character
wikipedia_df['other_source_deaths'] = wikipedia_df['other_source_deaths'].str.replace(r'\+', '')
```

```python
#removes commas
wikipedia_df['other_source_deaths'] = wikipedia_df['other_source_deaths'].str.replace(r'\,', '')
```

```python
wikipedia_df['other_source_deaths'].value_counts()
```

```python
wikipedia_df.iloc[1338] = wikipedia_df.iloc[1338].replace('601 (as of October 30 2011)', '601')
```

```python
wikipedia_df[wikipedia_df['other_source_deaths'] == '231000* 283000* 227898*']
```

```python
wikipedia_df.iloc[1237] = wikipedia_df.iloc[1237].replace('231000* 283000* 227898*', '283000')
```

```python
wikipedia_df[wikipedia_df['other_source_deaths'] == '26271 26000']
```

```python
wikipedia_df.iloc[1205] = wikipedia_df.iloc[1205].replace('26271 26000', '26000')
```

```python
wikipedia_df['other_source_deaths'] = wikipedia_df['other_source_deaths'].astype('float')
```

```python
wikipedia_df['country'].value_counts()
```

```python
#removes parenthesis and all characters contained
wikipedia_df['country'] = wikipedia_df['country'].str.replace(r'\(.+', '')
```

```python
wikipedia_df['country'].sort_values()
```

```python
wikipedia_df.iloc[1112]
```

```python
wikipedia_df.iloc[1260]
```

```python
#the spacing is causing issues so we'll remove it
wikipedia_df['country'] = wikipedia_df['country'].str.replace(r'\s', '')
```

```python
wikipedia_df['country'].value_counts()
```

```python
wikipedia_df[wikipedia_df['country'] == 'Iran,2005Qeshmearthquake']
```

```python
wikipedia_df[wikipedia_df['country'] == 'USTerritoryofAlaska']
```

```python
wikipedia_df[wikipedia_df['country'] == 'UnitedStatesMinorOutlyingIslands']
```

```python
wikipedia_df['country'].iloc[1258] = wikipedia_df['country'].iloc[1258].replace('Iran,2005Qeshmearthquake', 'Iran')
wikipedia_df['country'].iloc[378] = wikipedia_df['country'].iloc[378].replace('USTerritoryofAlaska', 'UnitedStates')
wikipedia_df['country'].iloc[560] = wikipedia_df['country'].iloc[560].replace('UnitedStatesMinorOutlyingIslands', 'UnitedStates')
wikipedia_df['country'].iloc[843] = wikipedia_df['country'].iloc[843].replace('UnitedStatesMinorOutlyingIslands', 'UnitedStates')
```

```python
wikipedia_df['country'].value_counts()
```

```python
wikipedia_df.info()
```

```python
#filling null values with NaN
wikipedia_df['lat'] = wikipedia_df['lat'].fillna('NaN')
wikipedia_df['long'] = wikipedia_df['long'].fillna('NaN')
wikipedia_df['depth_km'] = wikipedia_df['depth_km'].fillna('NaN')
wikipedia_df['magnitude'] = wikipedia_df['magnitude'].fillna('NaN')
wikipedia_df['secondary_effects'] = wikipedia_df['secondary_effects'].fillna('NaN')
```

```python
#filling null deaths with 0
wikipedia_df['pde_shaking_deaths'] = wikipedia_df['pde_shaking_deaths'].fillna(0)
wikipedia_df['pde_total_deaths'] = wikipedia_df['pde_total_deaths'].fillna(0)
wikipedia_df['utsu_total_deaths'] = wikipedia_df['utsu_total_deaths'].fillna(0)
wikipedia_df['em_dat_total_deaths'] = wikipedia_df['em_dat_total_deaths'].fillna(0)
wikipedia_df['other_source_deaths'] = wikipedia_df['other_source_deaths'].fillna(0)
```

```python
wikipedia_df.head(10)
```

```python
#create death column based on the highest of the 4 listed
wikipedia_df['death_total'] = wikipedia_df[['pde_shaking_deaths', 'pde_total_deaths', 
                                             'utsu_total_deaths', 'em_dat_total_deaths', 
                                             'other_source_deaths']].max(axis = 1)
```
