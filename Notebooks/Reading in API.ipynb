```python
import requests
import pandas as pd
from io import StringIO
import json 
import csv
```

```python
# capture data within the tn boarders with longitude and latitude and convert to csv and start the time from 1900 to now
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&minlatitude=35&maxlatitude=37&minlongitude=-86&maxlongitude=-84&starttime=1900-01-01&endtime=2020-10-17'
```

```python
r = requests.get(url)
```

```python
print(r.text[:1000])
```

```python
tn_quake_df = pd.read_csv(StringIO(r.text))
tn_quake_df.head()
```

```python
# capture the last word in the string wich is the state
tn_quake_df['state'] = tn_quake_df['place'].str[-9:]
```

```python
tn_quake_df['state'].unique()
```

```python
# remove data that is not Tennessee
tn_quake_df = tn_quake_df.groupby('state').get_group('Tennessee')
```

```python
# check that is correct
tn_quake_df.info()
```

```python
tn_quake_df.to_csv('cleaned_earthquake_data')
```
