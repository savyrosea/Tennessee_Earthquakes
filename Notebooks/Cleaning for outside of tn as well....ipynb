```python
import requests
import pandas as pd
from io import StringIO
import json 
import csv 
```

```python
# capture data within the tn boarders with longitude and latitude and convert to csv and start the time from 1900 to now
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&minlatitude=34&maxlatitude=37&minlongitude=-90.55&maxlongitude=-80&starttime=1900-01-01&endtime=2020-10-17'
```

```python
r = requests.get(url)
```

```python
print(r.text[:1000])
```

```python
outside_tn_quake_df = pd.read_csv(StringIO(r.text))
outside_tn_quake_df.head()
```

```python
# capture the last word in the string wich is the state
outside_tn_quake_df['state'] = outside_tn_quake_df['place'].str[-9:]

# capture the date  
outside_tn_quake_df['date'] = outside_tn_quake_df['time'].str[:10]

# capture the time  
outside_tn_quake_df['time'] = outside_tn_quake_df['time'].str[11:]

# capture the time  
outside_tn_quake_df['time'] = outside_tn_quake_df['time'].str[:8]
```

```python
outside_tn_quake_df.to_csv('../data/cleaned_outside_tn_earthquake_data.csv')
```
